```python
# Define function for string formatting of scientific notation
def sci_notation(num, decimal_digits=1, precision=None, exponent=None, math=False):
    """
    Returns a string representation of the scientific
    notation of the given number formatted for use with
    LaTeX or Mathtext, with specified number of significant
    decimal digits and precision (number of decimal digits
    to show). The exponent to be used can also be specified
    explicitly.
    """
    if exponent is None:
        exponent = int(floor(log10(abs(num))))
    coeff = round(num / float(10**exponent), decimal_digits)
    if precision is None:
        precision = decimal_digits

    ans = (r"{0:.{2}f}\times10^{{{{ {1:d} }}}}".format(coeff, exponent, precision) if math else
           r"${0:.{2}f}\times10^{{{1:d}}}$".format(coeff, exponent, precision))
    return ans

sci_notation(3.45e-3)

def stoich(x):
    return abs(x) if abs(x) !=1 else ""

def inc_or_decrease(val):
    return "increases" if val>0 else "decreases"
```

```python
# work for problem 4
def find_EA(k1, k2, t1, t2):
    T1 = t1 + 273.15
    T2 = t2 + 273.15
    EA = -R*log(k2/k1) / (1/T2 - 1/T1)
    A = k1 * exp(EA/(R*T1))
     
    return A, EA
```

```python
EAhere = 45150.2
Ahere = 17526.5
def get_k(t1, t2, A = Ahere, EA = EAhere):
    T1 = t1 + 273.15
    T2 = t2 + 273.15
    k1 = A*exp(-EA/(R*T1))
    k2 = A*exp(-EA/(R*T2))
    return array([k1, k2])
```

```python
# problem 7
kf1 = 4.8e3
kr1 = 3.6e3
kf2 = 1.3e-2
kf3 = 2.7e2

lbl_kf1 =  widgets.HTML(value = "<span style='color:blue'><b>k<sub>1</sub>: </b></span>")
lbl_kr1 =  widgets.HTML(value = "<span style='color:magenta'><b>k<sub>{-1}</sub>: </b></span>")
lbl_kf2 =  widgets.HTML(value = "<span style='color:green'><b>k<sub>2</sub>: </b></span>")
lbl_kf3 =  widgets.HTML(value = "<span style='color:red'><b>k<sub>3</sub>: </b></span>")

ent_kf1 = widgets.FloatLogSlider(value = kf1, min = 2, max = 4, step = 0.01, readout = True, 
                                continuous_update = False, readout_format = '.1e')
ent_kr1 = widgets.FloatLogSlider(value = kr1, min = 2, max = 4, step = 0.01, readout = True, 
                                continuous_update = False, readout_format = '.1e')
ent_kf2 = widgets.FloatLogSlider(value = kf2, min = -3, max = -1, step = 0.01, readout = True, 
                                continuous_update = False, readout_format = '.1e')
ent_kf3 = widgets.FloatLogSlider(value = kf3, min = 2, max = 4, step = 0.01, readout = True, 
                                continuous_update = False, readout_format = '.1e')
outbox7 = widgets.Output(layout = {'border': '1px solid black'})
ansbox7 = widgets.Output(layout = {'border': '1px solid black'})

toprow7 = widgets.HBox([lbl_kf1, ent_kf1, lbl_kr1, ent_kr1, lbl_kf2, ent_kf2, lbl_kf3, ent_kf3])

def change_problem7(change):
    outbox7.clear_output()
    ansbox7.clear_output()
    
    kf1 = ent_kf1.value
    kr1 = ent_kr1.value
    kf2 = ent_kf2.value
    kf3 = ent_kf3.value
    
    firstPart = r"The following 3-step mechanism has been proposed for the reaction of chlorine and chloroform:"
    eqn1 = r"\ce{Cl2 (g) <->[k_1][k_{-1}] 2 Cl (g)}"
    eqn2 = r"\ce{ Cl (g) + CHCl3 (g) ->[k_2] HCl (g) + CCl3 (g)}"
    eqn3 = r"\ce{Cl (g) + CCl3 (g) ->[k_3] CCl4 (g)}"
    
    secondPart = rf'''The numerical values of the rate constants for these steps are $k_1$ = <span style='color:blue'><b>
        {sci_notation(kf1)}</b></span>, $k_{{-1}}$ = 
        <span style='color:magenta'><b>{sci_notation(kr1)}</b></span>, 
        $k_2$ = <span style='color:green'><b>{sci_notation(kf2)}</b></span>, 
        and $k_3$ = <span style='color:red'><b>{sci_notation(kf3)}</b></span>. 
        Derive the rate law and the magnitude of the effective
        rate constant for the overall reaction.
        '''
    
    problem = rf'''
        The following 3-step mechanism has been proposed for the reaction of chlorine and chloroform:
        \par
        {{\centering
        \ce{{Cl2 (g) <-->[k_1][k_{-1}] 2 Cl (g)}} \par

        \ce{{ Cl (g) + CHCl3 (g) ->[k_2] HCl (g) + CCl3 (g)}} \par

        \ce{{Cl (g) + CCl3 (g) ->[k_3] CCl4 (g)}} \par
        }}
        The numerical values of the rate constants for these steps are $k_1$ = {sci_notation(kf1)}, $k_{{-1}}$ = 
        {sci_notation(kr1)}, $k_2$ = {sci_notation(kf2)}, and $k_3$ = {sci_notation(kf3)}.  Derive the rate law and the magnitude of the effective
        rate constant for the overall reaction.
        '''
    
    solution = r'''Since the rate constant $k_2$ is so much smaller than the other rate constants, it is clear that step two
    in the mechanism is the rate limiting step.  Thus, $$v = k_2 [Cl][CHCl_3]$$
    But $Cl$ is an intermediate, so it must be eliminated from the rate law.  To do this, 
    use the Rapid Equilibrium Approximation on step 1: 
    $$k_1[Cl_2] = k_{-1}[Cl]^2 \Rightarrow [Cl] = \left( \frac{k_1}{k_{-1}} [Cl_2] \right)^{1/2}$$
    Substitute this expression in place of $[Cl]$ to get the rate law:
    $$v = k_2 \sqrt{\frac{k_1}{k_{-1}}} [Cl_2]^{1/2} [CHCl_3] = k_{eff} [Cl_2]^{1/2} [CHCl_3]$$
    '''
    solution2 = rf'''Using the values given in the problem, $k_{{eff}} = \left( {sci_notation(kf2,math=True)} \right)
    \left( \frac{{ {sci_notation(kf1,math=True)} }}{{ {sci_notation(kr1,math=True)} }} \right)^{{1/2}} =$
    <span style='color:blue'><b>{sci_notation(kf2*sqrt(kf1/kr1))}</b></span>
    '''
    problems[6] = problem
    
    with outbox7:
        display(Markdown(firstPart))
        display(Math(eqn1))
        display(Math(eqn2))
        display(Math(eqn3))
        display(Markdown(secondPart))
        
    with ansbox7:
        display(Markdown(solution))
        display(Markdown(solution2))


for w in [ent_kf1, ent_kr1, ent_kf2, ent_kf3]:
    w.observe(change_problem7, names='value')
    
#problems[6] = problem
change_problem7('here')
widgets.VBox([toprow7, outbox7, ansbox7])
```
